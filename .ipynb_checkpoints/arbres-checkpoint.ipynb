```python
#import des bibliothèque

from pathlib import Path
import csv

#definition du chemin du fichier

csv_path = Path.cwd().parent /"I_data"

datafile = csv_path /"arbres.csv"
```

```python
#initialisation des en-têtes et des lignes dans des listes
rows = []
fields= []

#lecture du fichier csv

with open(datafile,'r') as csvfile :

#creation d'une variable lecture

    csvreader =csv.reader(csvfile)
#extraction des en-têtes

    fields = next(csvreader)
#extraction des lignes une par une
    for row in csvreader:
        rows.append(row)
        
    print(f"Total n. of rows: {csvreader.line_num}")
#affichage des en-têtes
    print(f"Field names are:{[fields for fields in fields]}")
```

```python
# extraction des deux premières lignes
with open(datafile, 'r') as csvfile: 
    linereader = csv.reader(csvfile, delimiter=',',quotechar='|')
    for row in list(linereader)[0:2]:
        print(' '.join(row))
```

```python
#extration des données de l'année utilisation dictionnaire
#affactation d'une liste de date
cropping_date = []


with open(datafile,'r') as csvfile :
#affectation des variables dans dictionnaire

    csv_dic_reader = csv.DictReader(csvfile)
#extration des lignes correspondant a l'annee de plantation
    for row in  csv_dic_reader:
        cropping_date.append(row['ANNEEDEPLANTATION'])
```

```python
print("les 50 première dates:\n",cropping_date[:60])
```

```python
print("les 50 dernières dates:\n",cropping_date[-50:])
```

```python
#compter les dates manquantes
count = cropping_date.count('')
print (count)
```

```python
#calcul de la moyenne d'arbre par annee
while("" in cropping_date) : 
    cropping_date.remove("")   
unique_dates_list = list(set(cropping_date))
unique_dates_list
```

```python
len(unique_dates_list)
```

```python
#calcul de la moyenne d'arbre par anneedate_occurence=[]
for date in unique_dates_list :
        date_occurence=cropping_date.count(date)
        print (date_occurence)
```

```python
#arbres post annee anniversaire
annee_anniversaire = '1985'
post_anniv=[]

with open(datafile,'r') as csvfile :
#affectation des variables dans dictionnaire

    csv_dic_reader = csv.DictReader(csvfile)
    
#extration des lignes correspondant a l'annee de plantation
    for row in  csv_dic_reader:
        if row['ANNEEDEPLANTATION']>=annee_anniversaire:
            post_anniv.append(row)
    print("nbr d'arbres plantée après année de naissance:", len(post_anniv))
```

```python
#arbres annee anniversaire
annee_anniversaire = '1985'
crop_anniv = []

with open(datafile,'r') as csvfile :
#affectation des variables dans dictionnaire

    csv_dic_reader = csv.DictReader(csvfile)
    
#extration des lignes correspondant a l'annee de plantation
    for row in  csv_dic_reader:
        if row['ANNEEDEPLANTATION']==annee_anniversaire:
            crop_anniv.append(row)
    
    print("nbr d'arbres plantée année de naissance:",len(crop_anniv))
```

```python
list_int=[]
for x in cropping_date:
    if x=='':
        cropping_date.remove('')
    else:
        list_int.append(int(x))
print(list_int)
```

```python

import matplotlib.pyplot as pyplot    
pyplot.title("arbres par année")
pyplot.hist(list_int,range(1920,2020),color=['red'],edgecolor='black',histtype='bar')
pyplot.xlabel("année")
pyplot.ylabel("nbr arbres")
pyplot.axis([1920,2020,0,5000])
pyplot.show()
```

```python
#extraction genre botanique
genre = []

with open(datafile,'r') as csvfile :
#affectation des variables dans dictionnaire

    csv_dic_reader = csv.DictReader(csvfile)
#extration des lignes correspondant a l'annee de plantation
    for row in  csv_dic_reader:
        genre.append(row['GENRE_BOTA'])
    print(len(genre))

```

```python
while("" in genre) : 
    genre.remove("")
print(len(genre))
genre_annee=list(zip(cropping_date,genre))
print(len(genre_annee))
```

```python

def takefirst(elem):
    return elem[0]

# sort list with key
genre_annee.sort(key=takefirst)

# print list
print('Sorted list:', genre_annee)
```
